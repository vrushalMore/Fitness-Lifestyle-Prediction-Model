```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_path = "../data/raw/innovize_final_ml.csv"
df = pd.read_csv(data_path)
```

# Display basic information

```python
print("Dataset Information:\n")
df.info()
print("\nSummary Statistics:\n", df.describe())
print("\nMissing Values:\n", df.isnull().sum())
```

# Visualize missing values

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cmap='viridis', cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```

# Distribution of numerical features

```python
numerical_features = ['phy_fitness', 'mindfulness', 'sleep_hrs', 'daily_avg_steps', 'daily_avg_calories']
df[numerical_features].hist(figsize=(12, 8), bins=20, edgecolor='black')
plt.suptitle("Distribution of Numerical Features")
plt.show()
```

# Relationship between physical fitness and mindfulness

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='phy_fitness', y='mindfulness', hue='is_healthy', palette='coolwarm', edgecolor='k')
plt.xlabel("Physical Fitness")
plt.ylabel("Mindfulness")
plt.title("Scatter Plot: Physical Fitness vs Mindfulness")
plt.show()
```

# Categorical feature distributions

```python
categorical_features = ['diet_pref', 'act_level', 'career', 'gender']
for feature in categorical_features:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=feature, palette='Set2')
    plt.title(f"Distribution of {feature}")
    plt.xticks(rotation=45)
    plt.show()
```
